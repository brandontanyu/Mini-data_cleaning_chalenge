```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avengers = pd.read_csv("avengers.csv", encoding='latin-1')
avengers.head()
```

```python
avengers['Year'].hist();
```

```python
avengers.info()
```

```python
avengers = avengers[avengers['Year'] >=1960]
```

```python
avengers['Year'].hist();
```

```python
death = [col for col in avengers.columns if 'Death' in col]
```

```python
df_deaths = avengers[death]

df_deaths.head()
```

```python
df_deaths.info()
```

```python
def total_deaths(row):
    num_deaths = 0
    columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
    for column in columns:
        death_value = row[column]
        if pd.isnull(death_value) or death_value == "NO":
            continue
        elif death_value == "YES":
            num_deaths += 1
    return num_deaths
```

```python
avengers['Deaths'] = avengers.apply(total_deaths,axis = 1)
```
